```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 9, 20)

y = x ** 3

z = x ** 2

figure = plt.figure()

axes = figure.add_axes([0,0,1,1])

axes.plot(x, z, label="Square Function")
axes.plot(x, y, label="Cube Function")
axes.legend()
```

```python
#PIE CHART PLOTTING

import matplotlib.pyplot as plt

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [150, 250, 100, 300]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
neth=plt.axes()
neth.legend(title="lang",loc="center")
neth.set_title("lang pie")
# Plot
plt.pie(sizes, labels=labels, colors=colors,shadow=True)
plt.show()
```
